# Bivariate Analysis of AIQ dataset

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('aqidataset.csv')
```

```python
df.head()
```

## 1. Bar plot (Numerical - Categorical)

```python
desired_states = ['NY', 'IL', 'FL']

df_filtered = df[df['state'].isin(desired_states)]

sns.barplot(x='state', y='Overall AQI Value', hue='Main Pollutant', data=df_filtered)
```

## 2. Boxplot (Numerical - Categorical)

```python
desired_states = ['NY', 'IL', 'FL']

df_filtered = df[df['state'].isin(desired_states)]

sns.boxplot(x='state', y='Overall AQI Value', hue='Main Pollutant', data=df_filtered)
```

## 4. Distplot (Numerical - Categorical)

```python
desired_states = ['NY', 'CA']
df_filtered = df[df['state'].isin(desired_states)]

sns.displot(data=df_filtered, x='Overall AQI Value', hue='AQI category', col='state', kde=True)
```

## 5. HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(df['Main Pollutant'],df['AQI category']))
```

## 6.ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(df['Main Pollutant'],df['AQI category']))
```
