```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('aqidataset.csv')
```

```python
df.head()
```

# 1. Categorical Data

a. Countplot

```python
sns.countplot(df['Main Pollutant'])
```

```python
sns.countplot(df['Site Name (of Overall AQI)'])

# there are many sites where AIQ was measured so y-axis label is overlapping
```

```python
sns.countplot(df['Source (of Overall AQI)'])
```

```python
sns.countplot(df['AQI category'])
```

b. Piechart

```python
df['Main Pollutant'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Source (of Overall AQI)'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['AQI category'].value_counts().plot(kind='pie',autopct='%.2f')
```

# 2. Numerical Data

a. Histogram

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Overall AQI Value'],bins=5)
```

```python
df['Ozone'] = df['Ozone'].astype(str)

plt.hist(df['Ozone'], bins=5)
plt.xlabel('Ozone Level')
plt.ylabel('Frequency')
plt.title('Histogram of Ozone Levels')

plt.xticks(rotation=30)
plt.gca().set_xticks(plt.gca().get_xticks()[::10])

plt.show()
```

b. Distplot

```python
sns.distplot(df['Overall AQI Value'])
```

c. Boxplot

```python
sns.boxplot(df['Overall AQI Value'])
```
